```python

f_capitulos1 = io.BytesIO(uploaded['f_capitulos (1).csv'])
f_capitulos1 = pd.read_csv(f_capitulos1, sep = ",", encoding="latin-1")
f_capitulos1.head(4)
```

```python
tabla_alcohol = pd.DataFrame({
    'edad': f_capitulos1['F_04'],
    'borrachera': f_capitulos1['F_09'],
    'cerveza': f_capitulos1['F_10_A'],
    'recordar': f_capitulos1['F_19'],
    "preocupacion": f_capitulos1["F_21"]
    })
tabla_alcohol.head(5)
```

```python
tabla_alcohol['cerveza'] = tabla_alcohol['cerveza'].fillna(0)
tabla_alcohol.head(5)
```

```python
tabla_alcohol['preocupacion'] = tabla_alcohol['borrachera'].fillna(2)
tabla_alcohol.head(5)
```

```python
tabla_alcohol1 = tabla_alcohol.dropna()
tabla_alcohol1.head(5)
```

```python
tabla_alcohol1['cerveza'] = tabla_alcohol1['cerveza'].astype(str)
tabla_alcohol1['recordar'] = tabla_alcohol1['recordar'].astype(str)
tabla_alcohol1['preocupacion'] = tabla_alcohol1['preocupacion'].astype(str)
```

```python
tabla_alcohol1.describe()
```

```python
tabla_alcohol['borrachera'] = tabla_alcohol['borrachera'].fillna(0)
tabla_alcohol.head(5)

tabla_alcohol['recordar'] = tabla_alcohol['recordar'].fillna(0) # 0 lo tomaremos para quienes no consumen alcohol
tabla_alcohol.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(tabla_alcohol1['edad'], bins=20, kde=True)
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(data=tabla_alcohol1, x='edad', hue='edad', palette='viridis', legend=False)
plt.title('Frecuencia de Personas por Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90) # Rotate x-axis labels if there are many ages
plt.grid(axis='y', alpha=0.75)
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

```python
plt.figure(figsize=(16, 8)) # Aumentar el tamaño del gráfico para mejor visibilidad
sns.countplot(data=tabla_alcohol1, x='cerveza', hue='cerveza', palette='Spectral', legend=False) # Cambiar la paleta de colores a 'Spectral'
plt.title('Distribución Detallada y Estilizada de Consumo de Cerveza', fontsize=16) # Aumentar el tamaño del título
plt.xlabel('Consumo de Cerveza (0 = No consume, otros valores = Cantidad)', fontsize=12) # Aumentar el tamaño de la etiqueta del eje X
plt.ylabel('Frecuencia', fontsize=12) # Aumentar el tamaño de la etiqueta del eje Y
plt.xticks(rotation=45, fontsize=10) # Rotar y ajustar tamaño de etiquetas del eje X
plt.yticks(fontsize=10) # Ajustar tamaño de etiquetas del eje Y
plt.grid(axis='y', alpha=0.6, linestyle='--') # Mejorar la rejilla

# Añadir etiquetas de valor en la parte superior de cada barra
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=9)

plt.tight_layout() # Ajustar el diseño para evitar que los elementos se superpongan
plt.show()
```
