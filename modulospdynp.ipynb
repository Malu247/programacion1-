```python
data = {
    'nombre': ['Ana', 'Luis', 'Carlos', 'Ana', None],
    'edad': [23, np.nan, 35, 23, 40],
    'salario': [1000, 2000, None, 1000, 5000],
    'ciudad': ['Bogotá', 'Medellín', 'Cali', 'Bogotá', 'Cali']
}
data
```

```python
data = pd.DataFrame(data)
data
```

```python
data.isnull().sum() #1 si falta valor, 0 si no falta ninguno
```

```python
data2 = data.dropna() #elimina las filas que tengan valores faltantes
data2
```

```python
data3 = data.dropna(subset=["salario"]) #solo se eliminan las filas de la columna escrita
data3
```

```python
data["edad"] = data["edad"].fillna(data["edad"].mean()) #En la columna edad, remplace los valores nulos por la media aritmetica de los valores de edad
data["nombre"] = data["nombre"].fillna("descoocido")
data
```

```python
data["ciudad"] = data["ciudad"].replace("Bogotá", "BOGOTA")
data
```

```python
dfn = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'nombre': [' Ana', 'LUIS ', 'Carlos', 'ana'],
    'edad': [23, 30, 35, 28],
    'salario': [1000, 2000, 3000, 1500],
    'ciudad': ['Bogotá', 'Cali', 'Bogotá', 'Medellín']
})
dfn
```

```python
dfn["salario_anual"] = dfn["salario"] * 12
dfn["edad_doble"] = dfn["edad"].apply(lambda x:x*2)
dfn
```

```python
dfn = dfn.rename(columns={"edad":"age"})
dfn
```

```python
grupo = dfn.groupby("ciudad")["salario"].mean()
grupo
```
