```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()
    password = input("Contraseña: ")

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    if usuarios[usuario] != password:
        raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

```python
productos = {
    "pan": 1.5,
    "leche": 2.0,
    "huevos": 3.0
}

print("Productos disponibles:", productos)
try:
  nombre = input("¿Qué producto quieres?: ").lower()
  if nombre not in productos:
    raise KeyError("El producto no está disponible")
  cantidad = int(input("Cantidad: "))
  if cantidad <= 0:
    raise ValueError("La cantidad debe ser mayor a 0")
  total = productos[nombre] * cantidad
  print("Total a pagar:", total)
except ValueError as e:
  print("Error de valor:", e)
except KeyError as e:
  print(e)
```

```python
import re
```

```python
texto = "Python es lenguaje"
resultado = re.search("Python", texto)
print(resultado)
```

```python
resultado = re.search("e", texto) #solo muestra la primera aparicion
print(resultado)
```

```python
texto = "tengo 23 perros y 3 gatos"
resultado = re.findall(r"\d", texto)
print(resultado) #El metodo fin all (con "\d") permite encontrar los digitos que aparecen en la cadena
```

```python
texto = "tengo 23 perros y 3 gatos"
resultado = re.findall(r"\d+", texto)
print(resultado) #El metodo fin all (con "\d") permite encontrar los digitos que aparecen en la cadena y unir los digitos contiguos
```

```python
texto = "Python no es divertido"
resultado = re.findall(r"\w+", texto)
print(resultado) #El metodo fin all (con "\w+") permite encontrar las letras que aparecen en la cadena y unir las contiguas para fromar palabras
```

```python
texto = "Si me gusta programar"
resultado = re.sub("Si","No",texto)
print(resultado)
```

```python
texto = "universidad"
resultado = re.findall(r"[aeiou]",texto)
print(resultado)
```

```python
texto = "Python Es Genial"
resultado = re.findall(r"[A-Z]",texto)
print(resultado)
```

```python
texto = "Hola mundo Python"
resultado = re.findall(r"\s", texto)
print(resultado)
```

```python
texto = "correos: ana@gmail.com y juan@hotmail.com"
resultado = re.findall(r"\w+@\w+\.\w+", texto)
print(resultado)
```
