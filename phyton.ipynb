```python
a, b = 0, 1 #El valorde b no esta actualizado
while b < 100:
  print(b)
  a, b = b, a+b
```

```python
a, b = 0, 1 #El valor de b esta actualizado
while b < 100:
  print(b)
  a = b
  b = a+b
```

```python
a, b = 0, 1
for b in range(20):   #Toma desde el 0 hasta el 19
    print(b)
    a, b = b, a + b
```

```python
x = int(input("Ingresa un entero  "))
if x % 3 == 0:
  print('Es multiplo de 3 :) ')
elif x % 3 == 1:
  print('No es multiplo de 3 porque el residuo es 1:( ')
else:
  print("No es multiplo de 3 porque el residuo es 2:( ")
```
