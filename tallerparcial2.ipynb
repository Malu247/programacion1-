```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, gaussian_kde

# Datos
data = df["nota"].dropna()

# Ajuste normal
mu, std = norm.fit(data)

# KDE
kde = gaussian_kde(data)

# Rango para graficar
x = np.linspace(data.min(), data.max(), 1000)

# Estilos
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# Histograma
sns.histplot(
    data,
    bins=30,
    stat="density",
    color="#4A90E2",
    alpha=0.4,
    edgecolor="black",
    label="Histograma"
)

# Curva normal
plt.plot(
    x,
    norm.pdf(x, mu, std),
    color="#E94E77",
    linewidth=2.5,
    label=f"Normal ajustada\nμ={mu:.2f}, σ={std:.2f}"
)

# KDE
plt.plot(
    x,
    kde(x),
    color="#2ECC71",
    linewidth=2.5,
    linestyle="--",
    label="KDE (Kernel Density)"
)

# Decoración
plt.title("Distribución de 'nota'\nHistograma, Normal ajustada y KDE",
          fontsize=16,
          fontweight="bold")

plt.xlabel("Nota", fontsize=13)
plt.ylabel("Densidad", fontsize=13)

plt.legend(fontsize=11)
plt.tight_layout()

# Mostrar
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, gaussian_kde

# ======================
# Datos
# ======================
data = df["nota"].dropna()

# ======================
# Ajustes
# ======================
mu, std = norm.fit(data)

# KDE con ajuste de bandwidth (puedes cambiar 0.8 -> más suave / 1.2 -> más rugoso)
kde = gaussian_kde(data, bw_method=0.8)

x = np.linspace(data.min(), data.max(), 1000)

# ======================
# Estilo visual
# ======================
sns.set_theme(style="whitegrid")
plt.figure(figsize=(13, 8))

# ======================
# Histograma
# ======================
sns.histplot(
    data,
    bins=30,
    stat="density",
    color="#5DADE2",
    alpha=0.35,
    edgecolor="white",
    linewidth=1
)

# ======================
# Normal ajustada
# ======================
normal_pdf = norm.pdf(x, mu, std)

plt.plot(
    x,
    normal_pdf,
    color="#C0392B",
    linewidth=3,
    label=f"Normal ajustada\nμ={mu:.2f}, σ={std:.2f}"
)

plt.fill_between(
    x,
    normal_pdf,
    color="#C0392B",
    alpha=0.15
)

# ======================
# KDE
# ======================
kde_values = kde(x)

plt.plot(
    x,
    kde_values,
    color="#27AE60",
    linewidth=3,
    linestyle="--",
    label="KDE (Kernel Density)"
)

plt.fill_between(
    x,
    kde_values,
    color="#27AE60",
    alpha=0.15
)

# ======================
# Líneas estadísticas
# ======================
plt.axvline(mu, color="black", linestyle=":", linewidth=2, label="Media")

plt.axvline(mu - std, color="gray", linestyle="--", linewidth=1.5)
plt.axvline(mu + std, color="gray", linestyle="--", linewidth=1.5)

# ======================
# Decoración
# ======================
plt.title(
    "Distribución de 'nota'\nHistograma, Normal ajustada y KDE",
    fontsize=18,
    fontweight="bold"
)

plt.xlabel("Nota", fontsize=14)
plt.ylabel("Densidad", fontsize=14)

plt.legend(frameon=True, shadow=True, fontsize=11)

# Quitar bordes superiores/derechos
sns.despine()

plt.tight_layout()

# ======================
# Guardar en alta calidad
# ======================
plt.savefig("distribucion_nota.png", dpi=300)
plt.savefig("distribucion_nota.svg")  # vectorial para papers

# Mostrar
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 7))

# Histograma A
sns.histplot(
    df["A"],
    bins=30,
    stat="density",
    color="#3498DB",
    alpha=0.4,
    edgecolor="white",
    label="Grupo A"
)

# Histograma B
sns.histplot(
    df["B"],
    bins=30,
    stat="density",
    color="#E74C3C",
    alpha=0.4,
    edgecolor="white",
    label="Grupo B"
)

# KDE opcional (suavizado)
sns.kdeplot(df["A"], color="#1F618D", linewidth=2.5)
sns.kdeplot(df["B"], color="#922B21", linewidth=2.5)

# Decoración
plt.title(
    "Comparación de distribuciones: Grupo A vs Grupo B",
    fontsize=17,
    fontweight="bold"
)

plt.xlabel("Valor", fontsize=13)
plt.ylabel("Densidad", fontsize=13)

plt.legend()
sns.despine()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(123)

grupo_A = np.random.normal(loc=4.0, scale=0.4, size=40)
grupo_B = np.random.normal(loc=4.3, scale=0.5, size=40)

df = pd.DataFrame({
    "A": grupo_A,
    "B": grupo_B
})

from scipy import stats
t_stat, p_val = stats.ttest_ind(df["A"], df["B"])

print("Estadístico t:", t_stat)
print("Valor p:", p_val)
```

Se rechaza que las calificaciones de ambos grupos sean iguales con una confianza de 95% porque el valor p es menor a 0,05
