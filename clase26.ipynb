```python
print("ADIVINA EL NÚMERO")
print("(Entre 1 y 5, escribe 0 para salir)")

numero_secreto = 3
numero_intentos = 2

while numero_intentos>0:

    intento = int(input("Adivina: "))
    if intento == 0:
        print("¡Hasta luego!")
        break
    if intento < 1 or intento > 5:
        print("Número fuera de rango, intenta de nuevo")
        continue
    if intento == numero_secreto:
        print("¡Correcto! Adivinaste")
        break
    else:
        if numero_intentos > 0:
            numero_intentos -= 1
            print(f"Quedan {numero_intentos} intentos")
        else:
            print("Te quedaste sin intentos")
```
