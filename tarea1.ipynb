Convertir una Lista

```python
lista = [1, 2, 3, 4, 5]
lista
```

```python
tupla = tuple(lista)
tuple
```

```python
lista2 = [("nombre", "James"), ("edad", 26)]
diccionario = dict(lista2) #Las listas simples no se pueden convertir en diccionarios
diccionario
```

```python
conjunto = set(lista)
conjunto
```

Convertir una Tupla

```python
tupla2 = ("Harry", "Niall", "Liam", "Zayn", "Louis")
tupla2
```

```python
lista3 = list(tupla2)
lista3
```

```python
diccionario2 = dict(enumerate(tupla2))
diccionario2 #Se utiliza el enumerate para cambiar una tupla simple
```

```python
conjunto2 = set(tupla2)
conjunto2
```

Convertir un Diccionario

```python
diccionario3 = {"nombre": "Justin", "apellido": "Bieber", "edad": 22 }
diccionario3
```

```python
lista4 = list(diccionario3.items()) #Al usar items se convierten en tuplas para poder hacer una lista
lista4
```

```python
tupla3 = tuple(diccionario3.items())
tupla3
```

```python
conjunto3 = set(diccionario3.items())
conjunto3
```

Convertir un Conjunto
