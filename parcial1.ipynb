```python
A.difference(B)
```

El metodo diference aplicado en conjuntos realiza una operación en la que se le resta a A la interseccion con el conjunto B. La respuesta, en este caso, es el conjunto de elementos que hacen parte de A y no hacen parte de B.

```python
A.symmetric_difference(B)
```

El método symmetric_difference aplicado a conjuntos devuelve la suma de los conjuntos resultantes al realizar la diferencia de A con B y de B con A. En este caso, la respuesta es el conjunto de elementos de A y B que no se encuentran en la interseccion entre ellos.

```python
def funcion1_parcial(m):
  try:
    m = float(m)
    if isinstance(m, int) and m >= 0:
      for n in range(0, (m+1)):
        if n == 1 or n == 0:
          print("El número es ", n)
        else:
          for x in range(2, n):
            if n % x == 0:
              print(n, 'es igual a', x, '*', n // x)
              break
          else:
            print(n, 'es un número primo')
    else:
      print("El número no es natural")
  except ValueError:
    print("El valor ingresado no es un número")
```

La funcion empieza por convertir el valor dado en float para verificar que es un número, si no se le puede aplicar significa que no es un numero y devuelve "El valor ingresado no es un numero". Si es un numero, la funcion continua por determinar si es 0 o 1 e imprime el numero es: (el numero ingresado), pero si no es ninguno de estos dos, la funcion evalua si es un numero primo intentado factorizarlo con 2, pero si no es posible se imprime "el numero no es primo". Finalmente, si no se encentra en el rango desde 0 hasta m+1 lo clasifica como un numero no natural.

```python
funcion1_parcial(-3)
```

```python
funcion1_parcial(6) #Ocurre un error
```

```python
funcion1_parcial("a")
```
