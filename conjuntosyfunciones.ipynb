```python
import math
```

```python
def volumen(a,h,c,d):
  return ((h*math.pi)/3)*((c*c)+(d*d)+(c*d))
```

```python
a = float(input("ingrese el lado a  "))
h = float(input("ingrese el lado h  "))
r_1 = float(input("ingrese el lado R1  "))
r_2 = float(input("ingrese el lado R2  "))

resultado = volumen(a,h,r_1,r_2)
print("el volumen es: ", resultado)
```

```python
def area(a,h,c,d):
  return (math.pi*((c*c)+(d*d)+(a*(c+d))))
```

```python
a = float(input("ingrese el lado a  "))
h = float(input("ingrese el lado h  "))
r_1 = float(input("ingrese el lado R1  "))
r_2 = float(input("ingrese el lado R2  "))

resultado = area(a,h,r_1,r_2)
print("el area es: ", resultado)
```

```python
def contador_vocales(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  contador1 = 0
  contador2 = 0
  contador3 = 0

  for letra in palabra.lower():
    if letra in vocales:
      contador1 += 1
    elif letra in consonantes:
      contador2 += 1
    else:
      contador3 += 1
  print("Cantidad de caracteres:", contador3)
  print("Cantidad de vocales:", contador1)
  print("Cantidad de consonantes:", contador2)
```

```python
def es_palindromo(cadena):
  cadena = cadena.lower()
  if cadena == cadena[::-1]:
    return "verdadero"
  else:
    return "falso"
```

```python
es_palindromo("camara")
```

```python
es_palindromo("reconocer")
```

```python
pal = input(str("introdice la palabra "))
es_palindromo(pal)
```

```python
def buscar(diccionario, clave):
  if clave in diccionario:
    return diccionario[clave]
  else:
    return "no existe"
```

```python
persona =  {"nombre": "Jose Alejandro","edad": 16, "icfes": 396, "estado": "calvo"}
print(buscar(persona, "nombre"))
print(buscar(persona, "correo"))
```
