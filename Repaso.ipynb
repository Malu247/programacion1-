```python
A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8, 9],
    [1, 2, 3]
]

# Creamos una matriz resultado con el mismo tamaño
resultado = []

# Recorremos filas
for i in range(len(A)):
    fila = []

    # Recorremos columnas
    for j in range(len(A[0])):
        suma = A[i][j] + B[i][j]
        fila.append(suma)

    resultado.append(fila)

# Mostrar resultado
print("La suma de las matrices es:")
for fila in resultado:
  print(fila)
```
