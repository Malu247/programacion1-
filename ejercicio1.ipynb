Ejercicio 1: Desarrolle un programa en que permita realizar operaciones entre números
complejos.

```python
def suma(z1, z2):
  return z1 + z2

def multiplicacion(z1, z2):
  return z1 * z2

def raiz(z, n):
  return z**(1/n)
```

```python
def pedir_entero():
  while True:
    try:
      return int(input("Ingrese un numero entero: "))
    except:
      print("Debe ingresar un numero entero")
```

```python
def pedir_opcion():
  while True:
    try:
      op = int(input("Ingrese una opcion:  "))

      if op in [1, 2, 3, 4]:
        return op
      else:
        print("Debe elegir una opcion del menu")
    except:
      print("Debe ingresar un numero valido")
```
