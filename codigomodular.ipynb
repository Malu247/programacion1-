```python
%%writefile src/productos.py
def pedir_productos():
  productos = []
  precios = []
  for i in range(int(input("Cuantos productos quieres indexar?  "))):
    nombre = input(f"ingrese nombre del producto {i+1}: ")
    precio = float(input(f"ingrese el precio del producto {i+1}: "))
    productos.append(nombre)
    precios.append(precio)
  glosario = dict(zip(productos,precios))
  return glosario

def mostrar_productos(glosario):
  for clave,valor in glosario.items():
    print(f"{clave}: {valor}")

def total_productos(glosario):
  total = 0
  for valor in glosario.values():
    total += valor
  return total

def total_iva(glosario):
  total = 0
  for valor in glosario.values():
    total += valor
  iva = total*0.19
  return iva
```

```python
%%writefile empresa/salario.py
def calcular_salario(horas, valor_hora):
  return horas * valor_hora

def calcular_bono(salario):
  if salario < 1000000:
    return salario * 0.1
  else:
    return 0
```

```python
%%writefile empresa/reportes.py
def mostrar_empleado(nombre, horas, valor_hora):
  print(f"Nombre: {nombre}")
  print(f"Horas trabajadas: {horas}")
  print(f"Valor por hora: {valor_hora}")

def mostrar_pago(salaraio, bono):
  print(f"Salario: {salaraio}")
  print(f"Bono: {bono}")
  print(f"Total a pagar: {salaraio + bono}")
```

```python
from empresa import registro
from empresa import salario
from empresa import reportes

nombre = registro.pedir_nombre()
horas = registro.pedir_horas()
valor = registro.pedir_valor_hora()

salario_base = salario.calcular_salario(horas, valor)
bono = salario.calcular_bono(salario_base)

reportes.mostrar_empleado(nombre, horas, valor)
reportes.mostrar_pago(salario_base, bono)
```
